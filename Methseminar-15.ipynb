```python
import sys
sys.path.append("/home/ahmedkashima/venv/lib/python3.12/site-packages")

import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

# Настройки модели
SPACE_SIZE = 250  # размер пространства
INITIAL_INFECTED = 10  # количество изначально зараженных
HEALTH_RANGE = (75, 150)  # диапазон здоровья
BOTTOM_RANGE = (30, 70)  # диапазон критического уровня здоровья
VELOCITY_RANGE = (1, 4)  # диапазон скоростей
```

```python
# Инициализация агентов

def init_agents(n_people=100, space_size=250):
    """
    Инициализация агентов:
    M - матрица позиций, скоростей и ускорений (n_people x 6)
    B - матрица состояний (жив, заражен, иммунитет) (n_people x 3)
    I - матрица здоровья и критического уровня (n_people x 2)
    """
    # Матрица позиций (x,y), скоростей (vx,vy) и ускорений (ax,ay)
    M = np.zeros((n_people, 6))
    M[:, :2] = np.random.rand(n_people, 2) * space_size  # начальные позиции
    M[:, 2:4] = (np.random.rand(n_people, 2) * (VELOCITY_RANGE[1] - VELOCITY_RANGE[0]) + VELOCITY_RANGE[0])
    
    # Матрица здоровья и критического уровня
    I = np.zeros((n_people, 2), dtype=int)
    I[:, 0] = np.random.randint(HEALTH_RANGE[0], HEALTH_RANGE[1]+1, n_people)  # здоровье
    I[:, 1] = np.random.randint(BOTTOM_RANGE[0], BOTTOM_RANGE[1]+1, n_people)  # критический уровень
    
    # Матрица состояний (alive, infected, immune)
    B = np.zeros((n_people, 3), dtype=bool)
    B[:, 0] = True  # все живы
    B[:INITIAL_INFECTED, 1] = True  # первые INITIAL_INFECTED заражены
    
    return M, B, I
```

```python
#  Функция обработки столкновений

def process_collision(M, B, idx1, idx2):
    """
    Обработка столкновения двух агентов:
    - Изменение направления движения
    - Возможная передача инфекции
    """
    # Вычисление угла между агентами
    dx = M[idx2, 0] - M[idx1, 0]
    dy = M[idx2, 1] - M[idx1, 1]
    alpha = np.arctan2(dy, dx)
    
    # Матрица поворота
    R = np.array([[np.cos(-alpha), -np.sin(-alpha)], 
                  [np.sin(-alpha), np.cos(-alpha)]])
    
    # Поворот скоростей, отражение и обратный поворот
    v_rot = R.dot(M[idx1, 2:4])
    v_rot_reflected = v_rot * np.array([1, -1])
    R_inv = np.array([[np.cos(alpha), -np.sin(alpha)], 
                      [np.sin(alpha), np.cos(alpha)]])
    v_new = R_inv.dot(v_rot_reflected)
    
    M[idx1, 2:4] = v_new
    
    # Передача инфекции (если idx2 заражен, а idx1 без иммунитета)
    if B[idx2, 1] and not B[idx1, 2]:
        if np.random.rand() < 0.5:  # или 0.3
            B[idx1, 1] = True
```

```python
# Основная функция итерации
def simulation_step(M, B, I, space_size):
    """
    Одна итерация моделирования:
    - Перемещение агентов
    - Обработка границ
    - Обновление здоровья
    - Проверка столкновений
    """
    dt = 0.1  # шаг времени
    
    # Индексы живых агентов
    alive_idx = np.where(B[:, 0])[0]
    
    for i in alive_idx:
        # Перемещение
        M[i, :2] += M[i, 2:4] * dt + 0.5 * M[i, 4:6] * dt**2
        
        # Обработка границ пространства
        for dim in [0, 1]:  # 0 - x, 1 - y
            if M[i, dim] < 0:
                M[i, dim] = -M[i, dim]
                M[i, 2+dim] *= -1
            elif M[i, dim] > space_size:
                M[i, dim] = 2*space_size - M[i, dim]
                M[i, 2+dim] *= -1
        
        # Обновление здоровья для зараженных
        if B[i, 1] and I[i, 0] > I[i, 1]:
            I[i, 0] -= 1
        
        # Проверка выздоровления
        if I[i, 0] <= I[i, 1]:
            B[i, 1] = False  # больше не заражен
            B[i, 2] = True    # теперь иммунитет
    
    # Проверка столкновений между всеми аг
```

```python
def plot_agents(M, B, step):
    plt.figure(figsize=(6, 6))
    healthy = B[:, 0] & ~B[:, 1] & ~B[:, 2]
    infected = B[:, 1]
    immune = B[:, 2]
    dead = ~B[:, 0]

    plt.scatter(M[healthy, 0], M[healthy, 1], color='blue', label='Healthy', s=10)
    plt.scatter(M[infected, 0], M[infected, 1], color='red', label='Infected', s=10)
    plt.scatter(M[immune, 0], M[immune, 1], color='green', label='Immune', s=10)
    plt.scatter(M[dead, 0], M[dead, 1], color='black', label='Dead', s=10)

    plt.title(f'Agents at Step {step}')
    plt.xlim(0, SPACE_SIZE)
    plt.ylim(0, SPACE_SIZE)
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Функция запуска моделирования

def run_simulation(n_people=100, n_iters=250, space_size=250):
    """
    Запуск полного моделирования
    Возвращает статистику по каждой итерации:
    - healthy: количество здоровых
    - infected: количество зараженных
    - immune: количество с иммунитетом
    - alive: количество живых
    """
    M, B, I = init_agents(n_people, space_size)
    
    stats = {
        'healthy': np.zeros(n_iters),
        'infected': np.zeros(n_iters),
        'immune': np.zeros(n_iters),
        'alive': np.zeros(n_iters)
    }
    
    for step in tqdm(range(n_iters), desc="Modeling"):
        simulation_step(M, B, I, space_size)
        
        # Сбор статистики
        stats['healthy'][step] = np.sum(B[:, 0] & ~B[:, 1] & ~B[:, 2])
        stats['infected'][step] = np.sum(B[:, 0] & B[:, 1])
        stats['immune'][step] = np.sum(B[:, 0] & B[:, 2])
        stats['alive'][step] = np.sum(B[:, 0])
    
    return stats
```

```python
# Задание 1 - 3 моделирования по 500 агентов
# Запуск 3 моделирований с разными random seed
plt.figure(figsize=(15, 10))

for i in range(3):
    np.random.seed(i)  # разные seed для разных результатов
    stats = run_simulation(n_people=500, n_iters=250)
    
    plt.subplot(3, 1, i+1)
    plt.plot(stats['healthy'], label='Healthy')
    plt.plot(stats['infected'], label='Infected')
    plt.plot(stats['immune'], label='Immune')
    plt.plot(stats['alive'], 'k--', label='Alive')
    plt.title(f'Simulation {i+1}')
    plt.xlabel('Iteration')
    plt.ylabel('Count')
    plt.legend()
    plt.grid()

plt.tight_layout()
plt.show()
```

```python
# Задание 2 - режим "не выходить из дома"

def run_simulation_with_lockdown(n_people=600, n_iters=300, lockdown_start=50, lockdown_duration=100):
    M, B, I = init_agents(n_people, SPACE_SIZE)
    
    stats = {
        'healthy': np.zeros(n_iters),
        'infected': np.zeros(n_iters),
        'immune': np.zeros(n_iters),
        'alive': np.zeros(n_iters)
    }
    
    # Сохраним оригинальные скорости
    original_velocities = M[:, 2:4].copy()
    
    for step in tqdm(range(n_iters), desc="Lockdown Modeling"):
        # Режим изоляции (50-150 итерации)
        if step == lockdown_start:
            M[:, 2:4] = 0  # останавливаем всех
        elif step == lockdown_start + lockdown_duration:
            M[:, 2:4] = original_velocities  # возвращаем скорости
        
        simulation_step(M, B, I, SPACE_SIZE)
        
        # Сбор статистики
        stats['healthy'][step] = np.sum(B[:, 0] & ~B[:, 1] & ~B[:, 2])
        stats['infected'][step] = np.sum(B[:, 0] & B[:, 1])
        stats['immune'][step] = np.sum(B[:, 0] & B[:, 2])
        stats['alive'][step] = np.sum(B[:, 0])
    
    return stats

# Запуск моделирования с изоляцией
stats = run_simulation_with_lockdown()

# Визуализация
plt.figure(figsize=(10, 6))
plt.plot(stats['healthy'], label='Healthy')
plt.plot(stats['infected'], label='Infected')
plt.plot(stats['immune'], label='Immune')
plt.plot(stats['alive'], 'k--', label='Alive')

# Отметки начала и конца изоляции
plt.axvline(50, color='r', linestyle='--', alpha=0.5, label='Lockdown start')
plt.axvline(150, color='g', linestyle='--', alpha=0.5, label='Lockdown end')

plt.title('Simulation with Lockdown (50-150 iterations)')
plt.xlabel('Iteration')
plt.ylabel('Count')
plt.legend()
plt.grid()
plt.show()
```

```python
# Задание 3 - введение вакцины

def run_simulation_with_vaccine(n_people=500, n_iters=250, vaccine_iter=30, vaccine_coverage=0.9):
    M, B, I = init_agents(n_people, SPACE_SIZE)
    
    stats = {
        'healthy': np.zeros(n_iters),
        'infected': np.zeros(n_iters),
        'immune': np.zeros(n_iters),
        'alive': np.zeros(n_iters)
    }
    
    for step in tqdm(range(n_iters), desc="Vaccine Modeling"):
        # Введение вакцины на 30-й итерации
        if step == vaccine_iter:
            # Выбираем 90% населения случайным образом
            vaccinated = np.random.choice(n_people, size=int(n_people*vaccine_coverage), replace=False)
            # Даем иммунитет (если не были уже иммунны)
            B[vaccinated, 2] = True
            # Если были заражены - вылечиваем
            B[vaccinated, 1] = False
        
        simulation_step(M, B, I, SPACE_SIZE)
        
        # Сбор статистики
        stats['healthy'][step] = np.sum(B[:, 0] & ~B[:, 1] & ~B[:, 2])
        stats['infected'][step] = np.sum(B[:, 0] & B[:, 1])
        stats['immune'][step] = np.sum(B[:, 0] & B[:, 2])
        stats['alive'][step] = np.sum(B[:, 0])
    
    return stats

# Запуск моделирования с вакциной
stats = run_simulation_with_vaccine()

# Визуализация
plt.figure(figsize=(10, 6))
plt.plot(stats['healthy'], label='Healthy')
plt.plot(stats['infected'], label='Infected')
plt.plot(stats['immune'], label='Immune')
plt.plot(stats['alive'], 'k--', label='Alive')

# Отметка введения вакцины
plt.axvline(30, color='purple', linestyle='--', alpha=0.5, label='Vaccine introduced')

plt.title('Simulation with Vaccine (introduced at iteration 30)')
plt.xlabel('Iteration')
plt.ylabel('Count')
plt.legend()
plt.grid()
plt.show()
```

```python
# Задание 4 - ограниченная начальная область

def init_agents_in_area(n_people=300, area_size=150, space_size=250):
    """Инициализация агентов в ограниченной области"""
    M = np.zeros((n_people, 6))
    # Начальные позиции в ограниченной области
    offset = (space_size - area_size) // 2
    M[:, :2] = np.random.rand(n_people, 2) * area_size + offset
    
    M[:, 2:4] = (np.random.rand(n_people, 2) * (VELOCITY_RANGE[1] - VELOCITY_RANGE[0]) + VELOCITY_RANGE[0])
    
    I = np.zeros((n_people, 2), dtype=int)
    I[:, 0] = np.random.randint(HEALTH_RANGE[0], HEALTH_RANGE[1]+1, n_people)
    I[:, 1] = np.random.randint(BOTTOM_RANGE[0], BOTTOM_RANGE[1]+1, n_people)
    
    B = np.zeros((n_people, 3), dtype=bool)
    B[:, 0] = True
    B[:INITIAL_INFECTED, 1] = True
    
    return M, B, I

def run_simulation_in_area(n_people=300, n_iters=200, area_size=150):
    M, B, I = init_agents_in_area(n_people, area_size, SPACE_SIZE)
    
    stats = {
        'healthy': np.zeros(n_iters),
        'infected': np.zeros(n_iters),
        'immune': np.zeros(n_iters),
        'alive': np.zeros(n_iters)
    }
    
    for step in tqdm(range(n_iters), desc="Area Modeling"):
        simulation_step(M, B, I, SPACE_SIZE)
        
        stats['healthy'][step] = np.sum(B[:, 0] & ~B[:, 1] & ~B[:, 2])
        stats['infected'][step] = np.sum(B[:, 0] & B[:, 1])
        stats['immune'][step] = np.sum(B[:, 0] & B[:, 2])
        stats['alive'][step] = np.sum(B[:, 0])
    
    return stats

# Запуск моделирования с ограниченной начальной областью
stats = run_simulation_in_area()

# Визуализация
plt.figure(figsize=(10, 6))
plt.plot(stats['healthy'], label='Healthy')
plt.plot(stats['infected'], label='Infected')
plt.plot(stats['immune'], label='Immune')
plt.plot(stats['alive'], 'k--', label='Alive')

plt.title('Simulation with Initial Agents in Limited Area (150x150)')
plt.xlabel('Iteration')
plt.ylabel('Count')
plt.legend()
plt.grid()
plt.show()
```
