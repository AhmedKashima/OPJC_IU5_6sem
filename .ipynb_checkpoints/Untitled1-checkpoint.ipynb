```python
lost_requests = np.array([5, 7, 6, 8, 6, 7, 5, 6, 7, 8])  
```

```python
# 1) Confidence probability for ε=1
n = len(lost_requests)
sample_mean = np.mean(lost_requests)
sample_std = np.std(lost_requests, ddof=1)
std_error = sample_std / np.sqrt(n)
epsilon = 1
t_value = epsilon / std_error
confidence_level = 2 * stats.t.cdf(t_value, df=n-1) - 1
print(f"1) Confidence probability for ε=1: {confidence_level:.4f}")

# 2) Required sample size for ε=0.25, P=0.99
epsilon = 0.25
confidence = 0.99
alpha = 1 - confidence
n_initial = 30
while True:
    t_critical = stats.t.ppf(1 - alpha/2, df=n_initial-1)
    required_n = (t_critical * sample_std / epsilon) ** 2
    if np.ceil(required_n) == n_initial:
        break
    n_initial = np.ceil(required_n)
print(f"2) Required sample size: {int(n_initial)}")
```

```python
print(f"Mean: {np.mean(data_b):.2f}")
print(f"Variance: {np.var(data_b, ddof=1):.2f}")
print(f"Std Dev: {np.std(data_b, ddof=1):.2f}")
print(f"Range: {np.max(data_b) - np.min(data_b)}")
print(f"Min: {np.min(data_b)}, Max: {np.max(data_b)}")
```

```python
def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    t = stats.t.ppf((1 + confidence) / 2, n-1)
    margin = t * std / np.sqrt(n)
    return (mean - margin, mean + margin)

for level in [0.95, 0.99, 0.999]:
    ci = confidence_interval(data_b, level)
    print(f"CI ({level:.3f}): ({ci[0]:.2f}, {ci[1]:.2f})")
```
